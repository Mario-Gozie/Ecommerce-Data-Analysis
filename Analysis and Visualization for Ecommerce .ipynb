# Importing Libraries I need

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
import seaborn as sns
from datetime import datetime
%matplotlib inline
```

# Data Cleaning

```python
df = pd.read_csv('Ecommerce_Data.csv', parse_dates= ['Date'])
```

```python
df.head()
```

```python
# I dont wish to continue with this unnamed column since I dont know what it is all about.  so let me select the columns I need.
df = df.loc[:,['Date', 'CustomerID', 'Country', 'InvoiceNo', 'StockCode', 'Description', 'Quantity', 'UnitPrice', 'Hour']]
```

```python
df.head()
```

```python
#Let me create some new columns and make better exploratory data analysis and visuals.
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].apply(lambda x: x.strftime('%B'))
df['Day'] = df['Date'].apply(lambda x : x.strftime('%A'))
#I need to create another column called revenue. it will have to be a multiplication of quantity and unit price column
df['Revenue'] = (df['Quantity'] * df['UnitPrice'])
```

```python
df.head()
```

```python
df.dtypes
```

```python
#looking at these data types, I need to work on the date, customerID,  and Hour.
# the Hour is supposed to be a string.
# customerID is supposed to be a string.
# date is supposed to have only date and exclude the time 
df['Date'] = pd.to_datetime(df['Date']).dt.date
df['CustomerID'] = df['CustomerID'].astype(str)
df['Hour'] = df['Hour'].astype(str)
df['Year'] = df['Year'].astype(str)
```

```python
df.dtypes
```

```python
df[df['Year']=='2010'].sort_values(by = 'Month') #we can see that data here is for only december of 2010
```

```python
df1 = df[df['Year']=='2010']
```

```python
#to confirm that business only took place in november for 2010
df1['Month'].unique()
```

```python
#Let me check for the total Revenue generated in dectmber 2010
```

```python
df1.groupby('Month').agg({'Revenue':'sum'})
```

```python
## I will like to know on which day of the week Highest sales was made.
df2 = df1.groupby('Day').agg({'Revenue':'sum'}). sort_values('Revenue', ascending= False) 
df2
```

```python
#Lets visualize what we have above
plt.style.use('ggplot')
fig, ax = plt.subplots()
ax.bar(df2.index, df2['Revenue'],color=plt.cm.cividis(0.2))
ax.set_xlabel('Days in December 2010', color = 'black')
ax.set_ylabel('Sum of Revenue', color = 'black')
ax.set_xticklabels(df2.index, rotation = 90)
plt.title('Budget per day for December 2010', color = 'black')
ax.tick_params(axis= 'y', color = 'black')
ax.tick_params(axis= 'x', color = plt.cm.cividis(0.2))
plt.show()
```
